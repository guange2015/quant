```python
gamblers = 100 #赌徒100

def casino(win_rate, win_once=1, loss_once=1,commission=0.01):
    my_money = 1000000
    play_cnt = 100000
    for _ in np.arange(0, play_cnt):
        w = np.random.binomial(1, win_rate)
        if w == 1:
            my_money += win_once
            my_money -= commission
        else:
            my_money -= loss_once
        if my_money<=0:
            break
    return my_money
```

```python
# 天堂赌场，没有老千，没有抽水
heaven_moneys = [casino(0.5, commission=0) for _ in range(gamblers)]
print(heaven_moneys)
plt.hist(heaven_moneys, bins=30);
```

```python
#有老千，导致胜率只有40%
heaven_moneys = [casino(0.4, commission=0) for _ in range(gamblers)]
print(heaven_moneys)
plt.hist(heaven_moneys, bins=30);
```

```python
#没有老千，有抽水
heaven_moneys = [casino(0.5, commission=0.01) for _ in range(gamblers)]
print(heaven_moneys)
plt.hist(heaven_moneys, bins=30);
```
