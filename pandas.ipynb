```python
import pandas as pd
import numpy as np
```

```python
index = pd.date_range('20180101',periods=8)
```

```python
df = pd.DataFrame(np.random.randn(8,4),index=index, columns=list('ABCD'))
```

```python
#转换
df.T
```

```python
#根据index排序
# axis 0 y轴   1 x轴
df.sort_index(ascending=False, axis=1)
```

```python
#根据值排序
# axis控制x,y轴
df.sort_values(by='B', axis=0)
```

### 选择数据

虽然可以直观的使用python下标取值，但更推荐用 `at`, `iat`, `loc`, `iloc`

```python
#直接访问列数据

df['A']
```

```python
# 切片
# 利用索引进行切片
df[0:3]
```

```python
# 利用索引名进行切片
df['2018-01-01':'2018-01-03']
```

### 使用label进行选择

```python
# 选择一行数据
df.loc['2018-01-01']
```

```python
# 选择多坐标值
# 参数逗号分隔，前一部分是行，后部分是列
df.loc[:, ['A', 'B']]
```

```python
df.loc['2018-01-01':'2018-01-02', ['A','B']]
```

```python
# 直接取标量
df.loc['2018-01-01', 'A']
```

```python
# at 有相同作用 ? 不行
df.at['2018-01-01', 'A']
```

### 通过position选择

* 第一个参数表示行
* 第二个参数表示列
* 省略而表示全取

参数三种形式
1. 纯数字定位
2. list，所有要取的数据
3. :大法，切片

```python
df.iloc[2]
```

```python
#多轴索引也可以
# 切片
df.iloc[3:5, 2:3]
```

```python
#python style
# 第一个参数list,可以把想要列都写出来
df.iloc[[1,2,7,6], 1]
```

### 通过布尔来选择

```python
# 数据筛选
df[df.A>0]
```

```python
# 将所有值进行对比，不同的为NaN
df[df>0]
```

```python
#isin
df[df['A'].isin([0.05808969972334659])]=1
df
```

### 合并数据

```python
df = pd.DataFrame(np.random.randn(10, 4))
df
```

```python
pieces = [df[:3], df[3:7], df[7:]]
pieces
```

```python
# 连接多个dataframe
pd.concat(pieces)
```

```python
# merge 
left = pd.DataFrame({'key': ['foo', 'foo', 'zz'], 'lval': [1, 2, 3]})
right = pd.DataFrame({'key': ['foo', 'foo'], 'rval': [4, 5]})
left.merge(right, on="key", how="left")
```

```python
df1 = pd.DataFrame({'Z':np.arange(0,10,1), 'K': np.random.randn(1,10)[0]})
```

```python
df2 = df.merge(df1,left_index=True, right_on='Z')
df2.drop('Z',axis=1)
```
