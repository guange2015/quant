```python
import os
import time

def match_file(file_path, exts=[]):
    for ext in exts:
        if file_path.endswith(ext):
            return True
    return False

def dir_path(folder):
    list = os.listdir(folder)
    for i in list:
        file_path = os.path.join(folder, i)
        if os.path.isfile(file_path):
            mtime = os.path.getmtime(file_path)
            mtime_struct = time.localtime(mtime)
            if mtime_struct.tm_year == 2018 and match_file(file_path, ['.js','.cpp','.h']):
                print("%s is file mtime %s" % (file_path, mtime_struct.tm_year))
        elif os.path.isdir(file_path):
            #print("%s is dir" % file_path)
            dir_path(file_path)
        else:
            pass
            #print("%s unkown" % file_path)
            
dir_path('/Users/guange/work/cocos2d/chaowan-game/frameworks')
```
